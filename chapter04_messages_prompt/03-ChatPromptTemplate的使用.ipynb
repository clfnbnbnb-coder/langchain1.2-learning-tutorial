# ChatPromptTemplate的使用

## 1、ChatPromptTemplate的实例化

### 1.1 from_messages() --推荐

举例：

```python
from langchain_core.prompts import ChatPromptTemplate

Chat_Prompt_Template = ChatPromptTemplate.from_messages([
    ("system","你是一个友好的AI助手，你的名字叫{name}"),
    ("user","你好，最近怎么样？"),
    ("ai","我很好，谢谢"),
    ("user","{user_input}"),
])

# 调用
result = Chat_Prompt_Template.invoke({"name":"卢本伟", "user_input":"2 + 2 = ？"})
print(result)
```

## 2、ChatPromptTemplate的调用

### 2.1 invoke()

传入参数：字典列表：返回值类型：ChatPromptValue

举例：

```python
from langchain_core.prompts import ChatPromptTemplate

Chat_Prompt_Template = ChatPromptTemplate.from_messages([
    ("system","你是一个友好的AI助手，你的名字叫{name}"),
    ("user","你好，最近怎么样？"),
    ("ai","我很好，谢谢"),
    ("user","{user_input}"),
])

# 调用
result = Chat_Prompt_Template.invoke({"name":"卢本伟", "user_input":"2 + 2 = ？"})
print(result)
print(type(result))
```

### 2.2 format()

参数类型：变量值：返回值类型：字符串

举例：

```python
from langchain_core.prompts import ChatPromptTemplate

Chat_Prompt_Template = ChatPromptTemplate.from_messages([
    ("system","你是一个友好的AI助手，你的名字叫{name}"),
    ("user","你好，最近怎么样？"),
    ("ai","我很好，谢谢"),
    ("user","{user_input}"),
])

# 调用
result = Chat_Prompt_Template.format(name = "卢本伟", user_input = "2 + 2 = ? ")
print(result)
print(type(result))
```

### 2.3 format_messages()

传入参数：变量值：返回值类型：消息列表

举例：

```python
from langchain_core.prompts import ChatPromptTemplate

Chat_Prompt_Template = ChatPromptTemplate.from_messages([
    ("system","你是一个友好的AI助手，你的名字叫{name}"),
    ("user","你好，最近怎么样？"),
    ("ai","我很好，谢谢"),
    ("user","{user_input}"),
])

# 调用
result = Chat_Prompt_Template.format_messages(name = "卢本伟", user_input = "2 + 2 = ? ")
print(result)
print(type(result))
```

## 3、结合大模型的调用

举例：

```python
from langchain.chat_models import init_chat_model
from langchain_core.prompts import ChatPromptTemplate
from dotenv import load_dotenv
import os

# 1、提供大模型
load_dotenv(override=True)
DASHSCOPE_API_KEY = os.getenv("DASHSCOPE_API_KEY")
DASHSCOPE_BASE_URL = os.getenv("DASHSCOPE_BASE_URL")

model = init_chat_model(
    model="qwen3.7-plus",
    model_provider="openai",
    # temperature=1.5,
    api_key=DASHSCOPE_API_KEY,
    base_url=DASHSCOPE_BASE_URL,
    # max_tokens=10,
)

# 2、提供提示词模板
Chat_Prompt_Template = ChatPromptTemplate.from_messages([
    ("system","你是一个友好的AI助手，你的名字叫{name}"),
    ("user","你好，最近怎么样？"),
    ("ai","我很好，谢谢"),
    ("user","{user_input}"),
])

# 调用
prompt_value = Chat_Prompt_Template.invoke({"name":"卢本伟", "user_input":"2 + 2 = ？"})
# prompt_value = result = Chat_Prompt_Template.format_messages(name = "卢本伟", user_input = "2 + 2 = ? ")

# 3、模型调用
response = model.invoke(prompt_value)
print(response)
```
